```python
import tkinter as tk
import pandas as pd

# Function that runs when button is clicked 
def verify_employee():
    name = name_entry.get()
    dept = dept_entry.get()

    try:
        data = pd.read_csv("C:/Users/user/Downloads/GIG-logistics(in).csv")

        match = data[
            (data["Name"].str.lower() == name.lower()) & 
            (data["Department"].str.lower() == dept.lower())
        ]

        if not match.empty:
            same_dept = data[data["Department"].str.lower() == dept.lower()]

            result = f"Welcome, {name} from {dept}!\n\nOther members:\n"

            for member in same_dept["Name"]:
                if member.lower() != name.lower():
                    result += f"- {member}\n"

            output_label.config(text=result)

        else:
            output_label.config(text="Sorry, employee does not exist.")

    except Exception as e:
        output_label.config(text=f"Error: {e}")

# Create main window 
window = tk.Tk()
window.title("Employee Verification System")
window.geometry("1500x500")

# 🎨 Set background color
bg_color = "#ADD8E6"  # light blue
window.configure(bg=bg_color)

# Name input 
tk.Label(window, text="Name", bg=bg_color).pack()
name_entry = tk.Entry(window)
name_entry.pack()

# Department input
tk.Label(window, text="Department", bg=bg_color).pack()
dept_entry = tk.Entry(window)
dept_entry.pack()

# Button 
tk.Button(window, text="Verify", command=verify_employee).pack()

# Output label 
output_label = tk.Label(window, text="", bg=bg_color)
output_label.pack()

# Run GUI
window.mainloop()
```
